```python
import pandas as pd
import numpy as np
#import plotly as pl
#from plotly.express import px
#from plotly.graph_objs import pg
from matplotlib import pyplot as plt
import seaborn as sea
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.nonparametric.kde import KDEUnivariate as kde
```

```python
df=pd.read_csv("c://j-works22/works/data/prac_kospi.csv")
x=df['dow']
y=df['kospi']
x=np.log(x)
y=np.log(y)

dy=np.diff(y)
# n=len(dy)
```

```python
# Density
sea.kdeplot(dy, shade=True)
```

```python
# CDF

kdensity=kde(dy)
kdensity.fit()
plt.plot(kdensity.cdf)
```
